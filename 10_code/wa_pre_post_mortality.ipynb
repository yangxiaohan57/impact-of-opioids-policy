```python
import pandas as pd
```

### Washington Pre-Policy Mortality

```python
# load mortality the data

death03 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2003.txt",
    error_bad_lines=False,
    sep="\t",
)
death04 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2004.txt",
    error_bad_lines=False,
    sep="\t",
)
death05 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2005.txt",
    error_bad_lines=False,
    sep="\t",
)
death06 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2006.txt",
    error_bad_lines=False,
    sep="\t",
)
death07 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2007.txt",
    error_bad_lines=False,
    sep="\t",
)
death08 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2008.txt",
    error_bad_lines=False,
    sep="\t",
)
death09 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2009.txt",
    error_bad_lines=False,
    sep="\t",
)
death10 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2010.txt",
    error_bad_lines=False,
    sep="\t",
)
death11 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2011.txt",
    error_bad_lines=False,
    sep="\t",
)
death12 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2012.txt",
    error_bad_lines=False,
    sep="\t",
)
death13 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2013.txt",
    error_bad_lines=False,
    sep="\t",
)
death14 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2014.txt",
    error_bad_lines=False,
    sep="\t",
)
death15 = pd.read_csv(
    "./US_VitalStatistics/Underlying Cause of Death, 2015.txt",
    error_bad_lines=False,
    sep="\t",
)
```

```python
# concatinate all pre-policy data

pre_policy = pd.concat(
    [death03, death04, death05, death06, death07, death08, death09, death10, death11],
    ignore_index=True,
)
```

```python
pre_policy
```

```python
# subset to Washington data and reformat County column

pre_policy[["County1", "State"]] = pre_policy.County.str.split(",", expand=True)
pre_policy_wa = pre_policy[pre_policy["State"] == " WA"]
pre_policy_wa.head()
```

```python
# drop columnes: Notes, County, County Code, Year Code; rename County1 to County

pre_policy_wa = pre_policy_wa.drop(
    ["Notes", "County", "County Code", "Year Code"], axis=1
)
pre_policy_wa = pre_policy_wa.rename(columns={"County1": "County"})
```

```python
pre_policy_wa.head()
```

```python
len(pre_policy_wa["County"].unique())
```

```python
# obtain codes for mortality for drug related causes

pre_policy_wa.groupby(
    ["Drug/Alcohol Induced Cause Code", "Drug/Alcohol Induced Cause"], as_index=False
).mean()
```

```python
# subset to mortality to drug related causes

pre_policy_wa_drug = pre_policy_wa[
    (pre_policy_wa["Drug/Alcohol Induced Cause Code"] == "D1")
    | (pre_policy_wa["Drug/Alcohol Induced Cause Code"] == "D2")
    | (pre_policy_wa["Drug/Alcohol Induced Cause Code"] == "D4")
    | (pre_policy_wa["Drug/Alcohol Induced Cause Code"] == "D9")
]
```

```python
pre_policy_wa_drug.head()
```

```python
# arrange by unique identifier: Year-State-County

pre_policy_wa_drug = pre_policy_wa_drug.groupby(
    ["Year", "State", "County"], as_index=False
).sum()
pre_policy_wa_drug.head()
```

```python
# convert year datatype from float to int

pre_policy_wa_drug["Year"] = pre_policy_wa_drug["Year"].astype(int)
pre_policy_wa_drug.head()
```

### Washington Post-Policy Mortality

```python
# concatinate all post-policy data

post_policy = pd.concat([death12, death13, death14, death15], ignore_index=True)
```

```python
post_policy
```

```python
# subset to Washington data

post_policy[["County1", "State"]] = post_policy.County.str.split(",", expand=True)
post_policy_wa = post_policy[post_policy["State"] == " WA"]
post_policy_wa.head()
```

```python
# drop columnes: Notes, County, County Code, Year Code; rename County1 to County

post_policy_wa = post_policy_wa.drop(
    ["Notes", "County", "County Code", "Year Code"], axis=1
)
post_policy_wa = post_policy_wa.rename(columns={"County1": "County"})
post_policy_wa.head()
```

```python
# subset to mortality to drug related causes

post_policy_wa_drug = post_policy_wa[
    (post_policy_wa["Drug/Alcohol Induced Cause Code"] == "D1")
    | (post_policy_wa["Drug/Alcohol Induced Cause Code"] == "D2")
    | (post_policy_wa["Drug/Alcohol Induced Cause Code"] == "D4")
    | (post_policy_wa["Drug/Alcohol Induced Cause Code"] == "D9")
]
post_policy_wa_drug.head()
```

```python
import warnings

warnings.filterwarnings("ignore")

# convert column Deaths to float datatype to prepare for sum in the next step

post_policy_wa_drug["Deaths"] = post_policy_wa_drug["Deaths"].astype(float)
```

```python
# arrange by unique identifier: Year-State-County

post_policy_wa_drug = post_policy_wa_drug.groupby(
    ["Year", "State", "County"], as_index=False
).sum()

post_policy_wa_drug.head()
```

```python
# convert year datatype from float to int

post_policy_wa_drug["Year"] = post_policy_wa_drug["Year"].astype(int)
post_policy_wa_drug.head()
```

### Merge Mortality Data with Population Data to Calculate Mortality Rate

```python
# load pre-policy population the data

wa_pop_09 = pd.read_excel("./Population/wa_pop_2000-2010.xls", header=[3])
wa_pop_19 = pd.read_excel("./Population/wa_pop_2010-2019.xlsx", header=[3])
```

```python
wa_pop_09.head()
```

```python
wa_pop_19.head()
```

#### Format Washington population data 2000-2009 and 2010-2019 before concatenate to one dataset, and subset to pre and post policy periods

#### Population data 2000-2009

```python
# keep columns with population data from 2003 to 2009; rename 'Unnamed: 0' to County

wa_pop_09 = wa_pop_09.drop(
    ["Unnamed: 1", 2000, 2001, 2002, "Unnamed: 12", "Unnamed: 13"], axis=1
)
wa_pop_09 = wa_pop_09.rename(columns={"Unnamed: 0": "County"})
wa_pop_09.head()
```

```python
# check last n rows with NaN values to determine the number of rows to drop

wa_pop_09.iloc[-9:]
```

```python
# remove first row - total state population and last 8 rows with NaN values

wa_pop_09.drop(wa_pop_09.head(1).index, inplace=True)
wa_pop_09.drop(wa_pop_09.tail(8).index, inplace=True)
wa_pop_09
```

#### Population data 2010-2019

```python
# keep columns with population data from 2010 to 2015; rename 'Unnamed: 0' to County

wa_pop_19 = wa_pop_19.drop(["Census", "Estimates Base", 2016, 2017, 2018, 2019], axis=1)
wa_pop_19 = wa_pop_19.rename(columns={"Unnamed: 0": "County"})
wa_pop_19.head()
```

```python
# check last n rows with NaN values to determine the number of rows to drop

wa_pop_19.iloc[-6:]
```

```python
# remove first row - total state population and last 5 rows with NaN values

wa_pop_19.drop(wa_pop_19.head(1).index, inplace=True)
wa_pop_19.drop(wa_pop_19.tail(5).index, inplace=True)
wa_pop_19
```

```python
# reformat County column to County name only

wa_pop_19[["County1", "State"]] = wa_pop_19.County.str.split(", ", expand=True)
wa_pop_19.head()
```

```python
# drop columnes: County, State

wa_pop_19 = wa_pop_19.drop(["County", "State"], axis=1)
wa_pop_19.head()
```

```python
# concatenate wa_pop_09 and wa_pop_09

wa_pop = pd.concat([wa_pop_09, wa_pop_19], axis=1)
wa_pop.head()
```

```python
# drop duplicate column -- County1

wa_pop = wa_pop.drop(["County1"], axis=1)
wa_pop.head()
```

```python
# remove the first character '.' in County Names

wa_pop["County"] = wa_pop["County"].str[1:]
wa_pop.head()
```

```python
# melt the data and arrange columns by unique identifier: Year-County

melted = wa_pop.melt(id_vars=["County"])
melted = melted.rename(columns={"variable": "Year", "value": "Population"})
melted = melted.groupby(["Year", "County"], as_index=False).sum()
melted
```

```python
# subset wa_pop for pre (before 2012) and post policy (starting 2012) periods

wa_pop_pre_policy = melted[melted["Year"] < 2012]
wa_pop_post_policy = melted[melted["Year"] >= 2012]
```

```python
# merge population data with pre policy drug mortality data

merged_pre = pre_policy_wa_drug.merge(wa_pop_pre_policy, how="left")
merged_pre
```

```python
# create county mortality rate column

merged_pre["Mortality Rate"] = merged_pre["Deaths"] / merged_pre["Population"]
merged_pre
```

#### Repeat the steps for post-policy population data

```python
# merge population data with post policy drug mortality data

merged_post = post_policy_wa_drug.merge(wa_pop_post_policy, how="left")
merged_post
```

```python
# create county mortality rate column

merged_post["Mortality Rate"] = merged_post["Deaths"] / merged_post["Population"]
merged_post
```
